### Word Vectors

This is an extension of the previous work, appling nlp. 

For this time, i will utilze Glove Vectors provided by Standford to create word vectors. 

Whats diff from one-hot encoding and bag-of-words models ?

one-hot encoding or bag-of words do not capture the contextual information from a word. Words are imported into ML without any contextual closeness. That is, those models do not capture syntactic and semantic relationships across collections of words.

In contrast, word vectors represent words as multidimensional continuous floating point numbers where semantically similar words are mapped to proximate points in geometric space.

```python
import pandas as pd
import numpy as np
from tqdm import tqdm
from nltk import word_tokenize
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
```

```python
embeddings_index = {}
f = open('data/glove.840B.300d.txt', encoding="utf-8")
for line in tqdm(f):
    values = line.split()
    word = values[0]
    coefs = np.asarray(values[1:]).astype(np.float)
    embeddings_index[word] = coefs
f.close()

print('Found %s word vectors.' % len(embeddings_index))
```

```python
train = pd.read_csv("data/train.zip")
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y= le.fit_transform(train.author.values)

from sklearn.model_selection import train_test_split

xtrain, xval, ytrain, yval = train_test_split(train.text.values,y,
                                             stratify=y, 
                                             random_state=42,
                                             test_size=0.1, 
                                             shuffle= True)
print(xtrain.shape)
print(xval.shape)
```

```python
def sent2vec(s):
    words = str(s).lower()
    words = word_tokenize(words)
    words = [w for w in words if not w in stop_words]
    words = [w for w in words if w.isalpha()]
    M = []
    for w in words:
        try:
            M.append(embeddings_index[w].astype(np.float))
        except:
            continue
    M = np.array(M)
    v = M.sum(axis=0)
    if type(v) != np.ndarray:
        return np.zeros(300)
    return v / np.sqrt((v ** 2).sum())
```

```python
xtrain_glove = [sent2vec(x) for x in tqdm(xtrain)]
xvalid_glove = [sent2vec(x) for x in tqdm(xval)]
```

```python
xtrain_glove = np.array(xtrain_glove)
xvalid_glove = np.array(xvalid_glove)
```

Applying XGboost 

```python
def multiclass_logloss(actual, predicted, eps=1e-15):
    """
    Multi class version of Logarithmic Loss metric.
    :param actual: Array containing the actual target classes
    :param predicted: Matrix with class predictions, one probability per class
    """
    # Convert 'actual' to a binary array if it's not already:
    if len(actual.shape) == 1:
        actual2 = np.zeros((actual.shape[0], predicted.shape[1]))
        for i, val in enumerate(actual):
            actual2[i, val] = 1
        actual = actual2

    clip = np.clip(predicted, eps, 1 - eps)
    rows = actual.shape[0]
    vsota = np.sum(actual * np.log(clip))
    return -1.0 / rows * vsota
```

```python
import xgboost as xgb

# test chunk
xg = xgb.XGBClassifier(nthread=10, silent=False)
xg.fit(xtrain_glove, ytrain)

pred = xg.predict_proba(xvalid_glove)

print ("logloss: %0.3f " % multiclass_logloss(yval, pred))
```

```python
#slight improvement in parameter
xg = xgb.XGBClassifier(max_depth=7, 
                       n_estimators=200, 
                       colsample_bytree=0.8,
                       subsample=0.8, 
                       nthread=10, 
                       learning_rate=0.1,
                       silent=False)

xg.fit(xtrain_glove, ytrain)
pred = xg.predict_proba(xvalid_glove)

print ("logloss: %0.3f " % multiclass_logloss(yval, pred))
```
