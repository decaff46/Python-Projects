```python
# dropping rows that every columns is na 
dat = dat[~dat.isna().all(axis = 1)]

dat.info()
```

```python
# removing white space
dat = dat.rename(columns=lambda x: x.strip())
dat.columns

#cols = ['name', 'status', 'market', 'funding_total_usd','equity_crowdfunding', 'country_code']
```

```python
#status_tab = dat['status'].value_counts().reset_index()
#status_tab = status_tab.rename(columns = {'status' : 'count','index':'status'})
status_tab = pd.crosstab(index=dat["status"], columns="count")
status_tab['%'] = 100 * status_tab / status_tab.sum()
status_tab['cum_%'] = 100 * status_tab['count'].cumsum()/status_tab['count'].sum()
status_tab
```

```python
fig = plt.figure(figsize=(7,7))

labels = dat['status'].value_counts().index.tolist()
sizes = dat['status'].value_counts().tolist()
explode = (0, 0, 0.2)
colors = ['#99ff99','#66b3ff','#ff9999']

plt.pie(sizes, 
        explode=explode, 
        labels=labels, 
        colors=colors, 
        autopct='%1.1f%%',
        shadow=False, 
        startangle=30)

plt.axis('equal')

plt.tight_layout()

plt.title('Startups current status', 
          fontdict=None, 
          position = [0.5,1],
          size = 'x-large'
         )

plt.show()
```

```python
#dat['funding_total_usd'] = dat['funding_total_usd'].astype('int')

# number seems to have whitespace too! 
dat = dat.apply(lambda x: x.str.strip() if x.dtype == 'object' else x)

# replace "," with ""
dat['funding_total_usd'] = dat['funding_total_usd'].str.replace(',', '')
# replace "-" with ""
dat['funding_total_usd'] = dat['funding_total_usd'].str.replace('-', '')
# replace " " with 0
dat['funding_total_usd'] = dat['funding_total_usd'].str.replace(' ', '')
dat['funding_total_usd'] = pd.to_numeric(dat['funding_total_usd'], errors='coerce')
```

```python
#dat.groupby('status').mean()['funding_total_usd']
fund_tab = pd.pivot_table(index=['status'], values=['funding_total_usd'], data=dat, aggfunc='mean')
fund_tab['%'] = 100 * fund_tab / fund_tab['funding_total_usd'].sum()
fund_tab
```

```python
from scipy import stats 
fig, ax = plt.subplots(1,2, figsize = (14,5))
sns.distplot(dat['funding_total_usd'].dropna(), ax = ax[0])  
stats.probplot(dat['funding_total_usd'].dropna(), dist="norm",fit=True, plot = ax[1])
            
plt.ylabel('Count')
plt.xlabel('Fnding (usd)')
plt.title("Distribution of total funding ", fontdict=None, position= [0.48,1.05], size = 'x-large')
plt.show()

print('Skewness :{:.3f}'.format(dat['funding_total_usd'].dropna().skew()))
print('Kurt : {:.3f}'.format(dat['funding_total_usd'].dropna().kurt()))
```

```python
from scipy.stats import norm
fig, ax = plt.subplots(1,2, figsize = (14,5))
sns.distplot(dat['funding_total_usd'].dropna(), ax = ax[0])  

dat['funding_total_usd_log'] = np.log1p(dat['funding_total_usd'])
sns.distplot(dat['funding_total_usd_log'].dropna(),
             fit = norm,
             rug=True, 
             rug_kws={"color": "g"},
             kde_kws={"color": "k", "lw": 3, "label": "KDE"},
             hist_kws={"histtype": "step", "linewidth": 3,"alpha": 1, "color": "g"}, 
            ax = ax[1])
            

plt.ylabel('Count')
plt.xlabel('Fnding (usd)')
plt.title("Distribution of total funding ", fontdict=None, position= [0.48,1.05], size = 'x-large')
plt.show()

print('Skewness :{:.3f}'.format(dat['funding_total_usd_log'].dropna().skew()))
print('Kurt : {:.3f}'.format(dat['funding_total_usd_log'].dropna().kurt()))
```

Still doesn't work due to the outliers! Let's apply 

```python
# 3 sigma
dat = dat[np.abs(dat["funding_total_usd"] - dat["funding_total_usd"].mean())<=(3*dat["funding_total_usd"].std())] 
```

```python


Q1 = dat['funding_total_usd'].quantile(0.25)
Q3 = dat['funding_total_usd'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = (Q1 - 1.5 * IQR)
upper_bound = (Q3 + 1.5 * IQR)

without_outlier = dat[(dat['funding_total_usd'] > lower_bound ) & (dat['funding_total_usd'] < upper_bound)]
```

```python
fig = plt.figure(figsize=(5,5))
plt.hist(without_outlier['funding_total_usd'].dropna(), bins=30,color = 'paleturquoise' )

plt.ylabel('Count')
plt.xlabel('Funding (usd)')
plt.title("Distribution of total funding ", fontdict=None, position= [0.48,1.05], size = 'x-large')
plt.show()
```
