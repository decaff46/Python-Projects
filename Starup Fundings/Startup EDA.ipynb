Still doesn't work due to the outliers! Let's apply 

```python
# 3 sigma
dat_threesig = dat[np.abs(dat["funding_total_usd"] - dat["funding_total_usd"].mean())<=(3*dat["funding_total_usd"].std())] 
```

```python
fig = plt.figure(figsize=(10,5))

plt.hist(dat_threesig['funding_total_usd'].dropna(), bins = 30, color='paleturquoise')
plt.ylabel('Count')
plt.xlabel('Funding (usd)')
plt.title("Distribution of total funding ", fontdict=None, position= [0.48,1.05], size = 'x-large')
plt.show()
```

```python
# IQR Method: 
Q1 = dat['funding_total_usd'].quantile(0.25)
Q3 = dat['funding_total_usd'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = (Q1 - 1.5 * IQR)
upper_bound = (Q3 + 1.5 * IQR)

dat_no_out = dat[(dat['funding_total_usd'] > lower_bound ) & (dat['funding_total_usd'] < upper_bound)]
```

```python
fig = plt.figure(figsize=(5,5))
plt.hist(dat_no_out['funding_total_usd'].dropna(), bins=30,color = 'paleturquoise' )

plt.ylabel('Count')
plt.xlabel('Funding (usd)')
plt.title("Distribution of total funding ", fontdict=None, position= [0.48,1.05], size = 'x-large')
plt.show()
```
