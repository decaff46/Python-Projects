Now that i know it works, I apply to the all!

```python
temp= []

for i in df.Translated_Review:
    text = re.sub("[^a-zA-Z]"," ",i) #removing non-characters
    text = text.lower() #lowering the case
    
    token = nltk.word_tokenize(text) #tokenizing the text
    le = nltk.WordNetLemmatizer() #foods --> food like 
    lemmatize = list(map(le.lemmatize, token)) #applying the lemmatize to all tokens
    string = ' '.join(lemmatize) # putting them together with one space to make it string
    temp.append(string) #putting the strings into temp list
    del text, token, lemmatize, string # for memory issue delete
```

```python
temp[:10]
```

Bag of Wods

```python
from sklearn.feature_extraction.text import CountVectorizer

max_features = 100

cv = CountVectorizer(max_features=max_features,
                    stop_words='english')

sparse_matrix = cv.fit_transform(temp).toarray()
sparse_matrix
```

```python
all_words=cv.get_feature_names()

print("Most used words Top 100: \n",all_words[:100])
```
