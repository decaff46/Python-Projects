```python
# import data: Starting point 
data = pd.read_csv('Twillio_data.csv')
data.head(5)
```

```python
# read benchmark 
benchmark = pd.read_csv('S&P Benchmark.csv')
benchmark.head()
```

```python
# functions 
def perct_change(previous, current): 
    '''Calculates the percentage change between two data frames '''
    result = (current - previous) / previous 
    return result * 100

def normalized(data): 
    '''Calculates the z score between two data sets'''
    mean = data.mean()
    sd = data.std()
    score = (data - mean) / sd
    return score 

def calc_beta(market,stock):
    covariance = np.cov(stock,market) # Calculate covariance between stock and market
    beta = covariance[0,1]/covariance[1,1]
    return beta
```

```python
# New Featues 
data['Change (%)'] = perct_change(previous= data['Open'], current=data['Close'])
benchmark['Change (%)'] = perct_change(previous= benchmark['Open'], current=benchmark['Close'])

# normalized data by volume 
data['Volume Normalized'] = normalized(data['Volume'])

# date stock 
data['Date'] = pd.to_datetime(data['Date'])
data['Day']  = data['Date'].dt.day
data['Month'] = data['Date'].dt.month
data['Weekday'] = pd.to_datetime(data['Date']).dt.weekday_name


# Range
data['Day Range'] = data['High'] - data['Low']
data['Cumulative Return'] = np.cumsum(data['Change (%)'])


count = []
count.append(0)
for i in data['Adj Close']: 
    count.append(i)
count.pop()

data['After hours Movement'] = data['Open'] - count

data = data.drop([0])
benchmark = benchmark.drop([0])

# Calculate twillio beta 
bencmark_return = benchmark['Change (%)']
stock_return = data['Change (%)']

benchmark_tally = []
stock_tally = []

count = 1
twillio_beta = []
for value in data['Change (%)']: 
    benchmark_tally = bencmark_return[1:count]
    stock_tally = stock_return[1:count]
    beta = calc_beta(benchmark_tally, stock_tally)
    twillio_beta.append(beta)
    count += 1 

    
data['Beta'] = twillio_beta
```

#### Beta  

A beta coefficient is a measure of the volatility, or systematic risk, of an individual stock in comparison to the unsystematic risk of the entire market.The beta calculation is used to help investors understand whether a stock moves in the same direction as the rest of the market, and how volatile or risky it is compared to the market. Beta can be used in conjution with return to indicate any anamoly in trading pattern. An increaed beta should coincide with an increased return, however where that pattern fails to converge should merit further inspeciton. Twilio trades at an average of 2x beta to the market however there are some instances where that ratio reaches as far as 12x, those anamoly would merir further investigation

#### After hours movement 
After hours movements captures movement in the stock price after the stock market has closed. 

**After hours Rationale:** 

Trading on Fresh Information: 
- Being able to trade after the normal markets close allows you to react quickly to breaking news stories or fresh information before the next day's market open.
- Pricing Opportunities: Although volatility is a risk associated with trading after hours, you may find some appealing prices during this time. 
- Convenience: Investors may prefer trading at off-peak times, and after-hours trading provides this added flexibility. 

Twilio generally doesn't move dramatically in the after hours.Changes outside of the norm should be investigated
